```python
!pip install ultralytics
from ultralytics import YOLO
import cv2
import random
import os
```

# Evaluate

```python
model = YOLO('./runs/detect/train5_Backup_11-12-2023/weights/best.pt') # change the model to which you want to evaluate
metrics = model.val(imgsz=960, data="./LogoDept_prep.yaml", device='cpu') # It run fine on colab, but runing on my PC using cuda got an error, so I use cpu instead.
metrics.box.map    # map50-95
metrics.box.map50  # map50
metrics.box.map75  # map75
metrics.box.maps   # a list contains map50-95 of each category
```

# Predict

```python
source = './datasets/LogoDet_prep/images/val'

results = model(source, save=True, imgsz=960)
```

```python
predict_imgs_path = os.listdir('./runs/detect/predict')
random_imgs_path = random.choices(predict_imgs_path, k=5)

window_name = 'preview'
for name in random_imgs_path:
    image = cv2.imread(os.path.join('./runs/detect/predict', name))
    cv2.imshow(window_name, image)
    cv2.waitKey()
    cv2.destroyAllWindows()
```
